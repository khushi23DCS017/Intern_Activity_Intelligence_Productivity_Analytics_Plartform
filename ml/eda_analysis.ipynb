# Exploratory Data Analysis (EDA)
Intern Activity Intelligence & Productivity Analytics Platform

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

# Load datasets
data_dir = r'e:\Intern Analysys'
activity_df = pd.read_csv(os.path.join(data_dir, 'synthetic_eod_logs.csv'))
python_df = pd.read_excel(os.path.join(data_dir, 'assignment_submissions_progress_Basic Python Programming.xlsx'))
sql_df = pd.read_excel(os.path.join(data_dir, 'assignment_submissions_progress_Basic SQL.xlsx'))
numpy_df = pd.read_excel(os.path.join(data_dir, 'assignment_submissions_progress_Data Processing using NumPy  Pa.xlsx'))
pyspark_df = pd.read_excel(os.path.join(data_dir, 'assignment_submissions_progress_Data Processing using Pyspark.xlsx'))

python_df['technology'] = 'Python'
sql_df['technology'] = 'SQL'
numpy_df['technology'] = 'NumPy'
pyspark_df['technology'] = 'PySpark'
assignments_df = pd.concat([python_df, sql_df, numpy_df, pyspark_df], ignore_index=True)
```

## Quick Data Overview

```python
display(activity_df.head())
display(assignments_df.head())
```

## Activity Distribution

```python
plt.figure(figsize=(10,6))
activity_df['Activity'].value_counts().head(10).plot(kind='bar', color='#3498db')
plt.title('Top 10 Activities by Frequency')
plt.xlabel('Activity')
plt.ylabel('Count')
plt.show()
```

## Technology Usage Patterns

```python
plt.figure(figsize=(8,5))
sns.countplot(data=assignments_df, x='technology', hue='technology', palette='viridis', legend=False)
plt.title('Technology Usage Patterns (Assignments)')
plt.xlabel('Technology')
plt.ylabel('Count')
plt.show()
```
